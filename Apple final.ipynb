```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('apple_quality.csv')
```

```python
df.head(5)
```

```python
# Histogram function
def create_histogram(data, column_name, bins=10):
    """
    Plots a histogram of the specified column in the dataset.

    Parameters:
    data (DataFrame): The dataset containing the data
    column_name (str): The name of the column to plot
    bins (int): Number of bins for the histogram
    """
    plt.hist(data[column_name], bins=bins, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.show()
    
#Histogram plot between data and size
create_histogram(df, 'Size')
```

```python
# Scatter plot function
def create_scatter_plot(data, x_column, y_column):
    """
    Plots a scatter plot between two specified columns.

    Parameters:
    data (DataFrame): The dataset containing the data
    x_column (str): The column for the x-axis
    y_column (str): The column for the y-axis
    """
    plt.scatter(data[x_column], data[y_column], color='green')
    plt.title(f'Scatter Plot of {x_column} vs {y_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()

# Scatter plot between Weight and Sweetness
create_scatter_plot(df.head(200), 'Weight', 'Sweetness')
```

```python
# Heatmap function
def create_heatmap(data):
    # Selecting only numeric columns
    """
    Plots a heatmap of the correlation matrix for the dataset's numeric columns.

    Parameters:
    data (DataFrame): The dataset containing the data
    """
    numeric_data = data.select_dtypes(include='number')
   
    plt.figure(figsize=(10, 8))
    sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('Correlation Heatmap')
    plt.show()
#Correlation heatmap of dataset
create_heatmap(df)
```
